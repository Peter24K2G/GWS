```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

# Subsetting the shapefile

```python
SZH = gpd.read_file("SHAPE/Zonificacion_hidrografica_2013.shp")
tAH = SZH.dissolve( by = "COD_AH")
TAH = tAH.reset_index()
AH = TAH.drop(columns=["OBJECTID_1","NOM_ZH","COD_SZH","COD_ZH","NOM_ZH","NOM_SZH","RULEID"],axis=1)
AH.to_file("Area_hidrografica.shp")
del AH
```

```python
AH = gpd.read_file("Area_hidrografica.shp")
plt.style.use("ggplot")
plt.figure(figsize=[20,20])
AH.plot()
plt.title("Áreas hidrográficas de Colombia")
plt.grid(which="minor")
plt.minorticks_on()
plt.show()
```
